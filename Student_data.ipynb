```python
import pandas as pd

data = pd.read_excel('PADRON-ELECTORAL-2023B-Provisional.xlsx')
data.head(2)
```

```python
#rows and columns
data.shape
# 124x8
```

```python
# Extract the desired column names without the prefix
col1_name = data.filter(like='Unnamed: 1').columns[0][9:]  # Extract characters after index 9
col2_name = data.filter(like='Unnamed: 2').columns[0][9:]
col3_name = data.filter(like='Unnamed: 3').columns[0][9:]  # Extract characters after index 9

# Concatenate the columns with a separator (optional)
data['Nombres'] = data['Unnamed: 1'].astype(str) + ' ' + data['Unnamed: 2'].astype(str)
data['Apellidos'] = data['Unnamed: 3'].astype(str)
data['Full Name'] = data['Nombres'] + ' ' + data['Apellidos']

data.head(3)
```

```python

data_filter = data[['LISTADO ESTUDIANTES MATRICULADOS GRADO','Nombres','Apellidos','Unnamed: 4','Unnamed: 5','Unnamed: 6']]
data_filter
```

```python
# Diccionario con los nombres antiguos y nuevos
nombres_columnas = {
    'LISTADO ESTUDIANTES MATRICULADOS GRADO': 'Estudiante',
    'Nombres': 'Nombre',
    'Apellidos': 'Apellidos',
    'Unnamed: 4': 'Cedula',
    'Unnamed: 5': 'Carrera',
    'Unnamed: 6': 'Modalidad',
}

# Rename columns
data_filter = data_filter.rename(columns=nombres_columnas)
# delete first row
data_filter = data_filter.drop(index=0)
data_filter.head(4)
```

```python
#Remove rows with at least one NaN value
data_filter_dropna = data_filter.dropna()
#count values
print(data_filter_dropna.count())

```

```python
# sort values

sort_data = data_filter.sort_values(['Nombre','Carrera'],ascending=[True,False])
sort_data.head(4)

```

```python
# Print info data
print(sort_data.info())
```

```python
# explore data for carrer

#TELECOMUNICACIONES

student_carrer = sort_data[(sort_data['Carrera'].isin(['TELECOMUNICACIONES','ELECTRÓNICA Y AUTOMATIZACIÓN']))]
student_carrer
```

```python

# Save the Data to a new CSV file named 'data_filtered.csv'

student_carrer.to_csv('data_filtered.csv', index=False)  # Optional arguments

print("Data filter saved to CSV file!")

```
